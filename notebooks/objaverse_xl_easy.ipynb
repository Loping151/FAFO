```python
import objaverse.xl as oxl
```

```python
annotations = oxl.get_annotations(
    download_dir="./data" # default download directory
)
annotations
```

Fetch your class easily:

```python
def check_dict(row, class_name):
    if class_name in row:
        return True
    return False

annotations_filtered = annotations[annotations['metadata'].apply(lambda row: check_dict(row, 'chair'))]

annotations_filtered
```

```python
oxl.download_objects(objects=annotations_filtered, download_dir='./data/')
```
